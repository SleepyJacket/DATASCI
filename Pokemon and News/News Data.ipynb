```python
import requests

URL="https://www.bmwblog.com/category/mini/?fbclid=IwAR3yIsmsOu6Ttz9bpmNouJF4TbLRLKfecR5x0VNWozdT9dIF3ebyLFFImUM"
```

```python
page=requests.get(URL)
print(page.content)
```

```python
from bs4 import BeautifulSoup

soup=BeautifulSoup(page.content, 'html.parser')
```

```python
bmw_content=soup.find(class_='post-listing')
news_item=bmw_content.find_all('article')
big_news=bmw_content.find('article', {'class':'bigArticle'})
news_item.remove(big_news)

#Check how many items there are
len(news_item)
```

```python
#Find needed information using first item found
url_to_full=news_item[0].find(class_="post-title").a.get('href')

fullpage=requests.get(url_to_full)
soup2=BeautifulSoup(fullpage.content, 'html.parser') 

date=soup2.find(class_="post-date").text.strip()
title=soup2.find(class_="post-title").text.strip()
content=soup2.find(class_='post-content').text.strip()
author=soup2.find(class_="byline").a.get('title')

print(url_to_full)

print(date)
print(title)
print(content)
print(author)
```

```python
#Find articles published on March 11, 2021 or March 12, 2021 only
for i in range(0, len(news_item)-1, 1): 
    url_to_full=news_item[i].find(class_="post-title").a.get('href')
    
    fullpage=requests.get(url_to_full)
    soup2=BeautifulSoup(fullpage.content, 'html.parser') 

    date=soup2.find(class_="post-date").text.strip()
    if(date == "March 11, 2021" or date == "March 12, 2021"):
        title=soup2.find(class_="post-title").text.strip()
        content=soup2.find(class_='post-content').text.strip()
        author=soup2.find(class_="byline").a.get('title')

        print(url_to_full)
        print(date)
        print(title)
        print(content)
        print(author)
```

```python
#Save to JSON
import json
```

```python
news_json = []

for i in range(0, len(news_item), 1): 
    if not(news_item[i].find(class_="bigArticle")):
    
        url_to_full=news_item[i].find(class_="post-title").a.get('href')
        fullpage=requests.get(url_to_full)
        soup2=BeautifulSoup(fullpage.content, 'html.parser') 

        date=soup2.find(class_="post-date").text.strip()
        if(date == "March 11, 2021" or date == "March 12, 2021"):
            title=soup2.find(class_="post-title").text.strip()
            content=soup2.find(class_='post-content').text.strip()
            author=soup2.find(class_="byline").a.get('title')

            news_json.append({
            "date": date,
            "title": title,
            "fullArticle": content,
            "author": author
        })
        
news_json

with open('news.json', 'w') as json_file:
    json.dump(news_json, json_file)
```
